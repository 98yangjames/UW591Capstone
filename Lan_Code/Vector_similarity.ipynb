```python
# store word embedding vectors in vectors
vectors =[]
  
def get_embeddings(word):
  inputs = tokenizer(word, return_tensors='pt')
  outputs = model(**inputs, output_hidden_states=True)
  last_hidden_states = outputs.hidden_states[-1]
  cls_tensor =last_hidden_states[0, 0, :]
  return cls_tensor.detach().numpy()

def get_token_embedding(names):
  for productname in names:
    vector = get_embeddings(productname)
    vectors.append(vector)
```

```python
# load embedding data and product name
product_name_vectors = np.load('/content/drive/MyDrive/product_name_embeddings.npy')
```

```python
words =[]
with open('/content/drive/MyDrive/unique_product_name.txt', 'rb') as f:
  for line in f:
    word = line.strip().decode('utf-8')
    words.append(word)
```

```python
def cos_cdist(vector, matrix=product_name_vectors):
  # comput the cosine distances between each row of matrix and vector
  # 1 - Cosine_Distance = Cosine_Similarity
  v = vector.reshape(1, -1)
  return scipy.spatial.distance.cdist(matrix, v, 'cosine').reshape(-1)
```

```python
def sort_cosine_value(cosine_vals):
  # return the index of closed product name
  return (cosine_vals).argsort()
```

```python
def get_x_closed_words(word, num =1):
  embedding = get_embeddings(word)
  cosine_distances = cos_cdist(embedding)
  
  words_index = sort_cosine_value(cosine_distances)
  
  closed_w =[]
  for w in words_index[:num]:
    cosine_similarity = 1 - cosine_distances[w]
    
    if cosine_similarity >=0.8:
      closed_w.append((words[w], cosine_similarity ))
  
  return closed_w
```

```python
get_x_closed_words('dompone', num =5)
```
